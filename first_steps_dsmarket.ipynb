```python
def rellenar_secuencia(series):
    """
    Rellena NaNs en una serie sumando 1 consecutivamente 
    a partir del ltimo valor vlido.
    """
    # 1. Rellenar NaNs con el ltimo valor vlido
    # [11, 12, 13, NaN, NaN] -> [11, 12, 13, 13, 13]
    relleno_base = series.ffill()
    
    # 2. Crear un contador que solo incremente en los NaN originales
    # [11, 12, 13, NaN, NaN] -> isnull() -> [F, F, F, T, T]
    # .cumsum() -> [0, 0, 0, 1, 2]
    incremento = series.isnull().cumsum()
    
    # 3. Sumar la base + el incremento
    # [11, 12, 13, 13, 13] + [0, 0, 0, 1, 2] = [11, 12, 13, 14, 15]
    return relleno_base + incremento

# Aplicar la funcin a cada grupo
pd_prices_full_week = pd_prices.groupby(['item', 'store_code'])['yearweek'].apply(rellenar_secuencia).reset_index().set_index('level_2')

pd_prices_full_week.sort_index(inplace=True)

pd_prices['yearweek'] = pd_prices_full_week['yearweek']

pd_prices_full_week.yearweek.isnull().sum()

### Ultima fecha de week: 201617.0
### A partir de esa fecha, datos para predecir
```

### Feature Engineering

```python
df_weekly.head()
```

```python
# Lags (Past values)
# Previous week sales (t-1)
df_weekly['lag_1_sale'] = df_weekly.groupby(['item', 'store_code'])['sale'].shift(1)
df_weekly['lag_1_sellprice'] = df_weekly.groupby(['item', 'store_code'])['sell_price'].shift(1)

# Sales 4 weeks ago (approx 1 month)
df_weekly['lag_4_sale'] = df_weekly.groupby(['item', 'store_code'])['sale'].shift(4)
df_weekly['lag_4_sellprice'] = df_weekly.groupby(['item', 'store_code'])['sell_price'].shift(4)

# Sales 52 weeks ago (approx 1 year)
df_weekly['lag_52_sale'] = df_weekly.groupby(['item', 'store_code'])['sale'].shift(52)
df_weekly['lag_52_sellprice'] = df_weekly.groupby(['item', 'store_code'])['sell_price'].shift(52)

# Variation / Detrending (Target for stationary models)
# "Diff": Absolute difference (Current - Previous)
df_weekly['sales_diff'] = df_weekly.groupby(['item', 'store_code'])['sale'].diff(1)
df_weekly['price_diff'] = df_weekly.groupby(['item', 'store_code'])['sell_price'].diff(1)

# Rolling Features (Smoothing for Power BI visualization)
# 4-week moving average (shows trend without noise)
df_weekly['rolling_mean_4w_sale'] = df_weekly.groupby(['item', 'store_code'])['sale'].transform(lambda x: x.rolling(window=4).mean())
df_weekly['rolling_mean_4w_sellprice'] = df_weekly.groupby(['item', 'store_code'])['sell_price'].transform(lambda x: x.rolling(window=4).mean())

# Cleanup
df_weekly.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set a consistent theme for all plots
sns.set_theme(style="whitegrid")

def plot_aggregated_trend(df, group_col, date_col='yearweek', sales_col='sale'):
    """
    Plots sales trends over time, grouped by a specified dimension (e.g., category, region).

    Args:
        df (pd.DataFrame): The weekly aggregated DataFrame.
        group_col (str): The column to group by (e.g., 'category_x', 'region', 'store_code').
        date_col (str): The column containing date information.
        sales_col (str): The column containing the sales metric to plot.
    """
    try:
        # Aggregate data for plotting
        plot_data = df.groupby([group_col, date_col])[sales_col].sum().reset_index()
        
        plt.figure(figsize=(16, 7))
        sns.lineplot(data=plot_data, x=date_col, y=sales_col, hue=group_col, palette='deep')
        
        plt.title(f'Weekly {sales_col.title()} Trend by {group_col.title()}', fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel(f'Total Weekly {sales_col.title()}', fontsize=12)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.plot()
        
        # Save the plot
        #filename = f'weekly_sales_trend_by_{group_col}.png'
        #plt.savefig(filename, bbox_inches='tight')
        #print(f"Plot saved as '{filename}'")
        #plt.clf() # Clear the figure to avoid overlap in notebooks
        
    except Exception as e:
        print(f"Error plotting aggregated trend: {e}")

def plot_top_n_performers(df, entity_col, sales_col='sale', n=20, group_colors='region'):
    """
    Creates a horizontal bar chart of the top N performers (e.g., items, stores).

    Args:
        df (pd.DataFrame): The weekly aggregated DataFrame.
        entity_col (str): The column to analyze (e.g., 'item', 'store').
        sales_col (str): The sales metric to sum up.
        n (int): The number of top performers to display.
    """
    gb = [entity_col, group_colors]
    try:
        # Calculate total sales for the entity
        top_n_data = df.groupby(gb)[sales_col].sum().nlargest(n).reset_index()
        
        plt.figure(figsize=(12, n * 0.4))  # Dynamic height based on N
        sns.barplot(
            data=top_n_data,
            y=entity_col,
            x=sales_col,
            orient='h',
            hue=group_colors
        )
        
        plt.title(f'Top {n} {entity_col.title()} by Total {sales_col.title()}', fontsize=16)
        plt.xlabel(f'Total {sales_col.title()}', fontsize=12)
        plt.ylabel(entity_col.title(), fontsize=12)
        plt.plot()
        
        # Save the plot
        #filename = f'top_{n}_{entity_col}_by_{sales_col}.png'
        #plt.savefig(filename, bbox_inches='tight')
        #print(f"Plot saved as '{filename}'")
        #plt.clf()
        
    except Exception as e:
        print(f"Error plotting top N performers: {e}")

```
